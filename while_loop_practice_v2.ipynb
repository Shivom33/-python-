Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here
n = int(input("Enter a number: "))
rev = 0

while n > 0:
    a = n % 10
    rev = rev * 10 + a
    n = n // 10

print("Reversed number:", rev)
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
n = int(input("Enter a number: "))
count = 0
while n > 0:
    n = n // 10
    count += 1

print("Number of digits:", count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
n = int(input("Enter a number: "))
original = n
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if rev == original:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```
